```python
import twint
import nest_asyncio 
nest_asyncio.apply()

c = twint.Config()
c.Limit = 500
c.Search = "Ganjar Pranowo" # topic
c.Pandas = True
c.Store_csv = True
twint.run.Search(c)
Tweets_df = twint.storage.panda.Tweets_df
```

```python
# Tweets_df.columns
```

```python
data = pd.DataFrame(Tweets_df, columns=["username", "name", "tweet"])
```
